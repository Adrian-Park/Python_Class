# 이메일 전달(forward)
## 이메일 한개 가져온다
## 가져온 이메일의 본문 내용을 읽어서 전달할 메시지로 구성          제목, 발신자 정보 설정 등
## 메일 보내기

```python
import imapclient
import email
from email.header import decode_header, Header
from datetime import date, timedelta

imap = imapclient.IMAPClient('imap.gmail.com', ssl = True)
imap.login('my_mail', 'password') #id, password
imap.select_folder('INBOX', readonly=False)
uids = imap.search(['ON', date.today()])[:1] #오늘 받은 메일 한개 가져오기

rmsgs = imap.fetch(uids, ['BODY[]'])
m = rmsgs[2024][b'BODY[]'] #uid는 검색된 메일 id값
message = email.message_from_bytes(m) #전달할 이메일의 메시지로 구성

message.get('subject')
subject, encoding = decode_header(message.get('subject'))[0]
str(subject)

message['from']
decode_header(message['from'])

s = ''
for text, encoding in decode_header(message['from']):
    if not encoding:
        encoding = 'utf-8'
    s += str(text)
s

new_subject = 'Fwd({}):{}'.format(s, str(subject))
new_subject
```

```python
from_addr ='my_mail'
to_addr = 'password'
message.replace_header('from', from_addr)
message.replace_header('to', to_addr)
del message['cc']
del message['bcc']
message.replace_header('subject', Header(new_subject.encode('utf-8'), 'utf-8'))

import smtplib
smtp = smtplib.SMTP('smtp.gmail.com', 587)
smtp.starttls()
smtp.login('my_mail', 'password')
smtp.sendmail(from_addr, to_addr, message.as_bytes())
smtp.quit()
```

```python
import sys
import sched

def my_job(text="..."):
    print(text)
    sys.stdout.flush()

#블로킹 모드로 실행
s = sched.scheduler()
s.enter(5, 1, my_job) #5초 후 my_job() 호출 실행
s.enter(10, 1, my_job, argument=('good!',))
#10초 후 my_job() 호출 실행
s.run() # Blocking
print('done')
```

```python
import time
s = sched.scheduler()
s.enterabs(time.time()+5, 1, my_job)
```

```python
import threading
import time

def my_job(text):
    print(text)
    
threading.Timer(5, my_job, ('good!!',)).start() #5초 대기 후 실행
threading.Timer(5, my_job, ('bad!!',)).start() #5초 대기 후 실행
print('other works continued..')
time.sleep(5)
```

```python
import threading
import time

#지정된 시간에 인수로 전달받은 함수를 실행시키는 함수
def run_it_at(at, func, args=()):
    t = time.strptime(at, '%Y-%m-%d %H:%M:%S')
    t = time.mktime(t) # 지정된 시간을 숫자 시간으로 변환
    diff_sec = t - time.time() # 현재 시간과의 차이 => 대기시간(초)
    print(diff_sec)
    threading.Timer(diff_sec, func, args).start()
    
run_it_at('2021-04-22 15:21:00', my_job, ('excellent',))
print('other works continued..')
time.sleep(5)
#time.sleep()을 해주는 이유 : 메인스레드가 다른 자식 스레드들이 종료될 때까지 기다려주기 위해.
```

```python
import schedule
import time
def job(msg="I'm working..."):
    print(msg)

schedule.every().seconds.do(job) #1초마다
schedule.every(5).seconds.do(job, ('5 sec',)) #매 5초마다
schedule.every(10).minutes.do(job) #10분마다
schedule.every().hour.do(job, ('hour',)) #매시간
schedule.every().day.at("10:30").do(job)
schedule.every(5).to(10).seconds.do(job) #5초에서 10마다(랜덤으로 선택)
schedule.every().wednesday.at("13:15").do(job) #수요일 13:15마다
```

## 연습문제
### 뉴스리더 : 1시간 간격으로 뉴스 url 수집해서 메일로 보내기

```python
import smtplib
import feedparser
from email.mime.text import MIMEText
from email.header import Header

#관심 RSS feed 주소를 추가, 변경한다.
rss_feeds = [
    'http://file.mk.co.kr/news/rss/rss_30100041.xml',
    'http://file.mk.co.kr/news/rss/rss_50300009.xml'
]
#words of interest 관심 단어
WOI = ['금리', '환율']

out = []
for feed in rss_feeds:
    d = feedparser.parse(feed)
    for entry in d.entries:
        for w in WOI: #관심 단어가 제목이나 요약에 있는지 살핀다
            if w in entry['title'] or w in entry['summary']:
                s = '* {} {}'.format(entry['title'], entry['link'])
                out.append(s)
                break
message = '\n'.join(out)

if message:
    subject = 'RSS News'
    mail_from = '보내는 사람 메일 주소'
    mail_to = '받는 사람 메일 주소'
    id_ = 'smtp 서버에 로그인 가능한 메일 id'
    pw_ = 'smtp 서버에 로그인 가능한 메일 password'
    
    smtp = smtplib.SMTP('smtp.gmail.com', 587)
    smtp.ehlo()
    smtp.starttls()
    smtp.login(id_, pw_)
    
    msg = MIMEText(message.encode('utf-8'), _subtype='plain', _charset='utf-8')
    msg['Subject'] = Header(subject.encode('utf-8'), 'utf-8')
    msg['From'] = mail_from
    msg['To'] = mail_to
    
    smtp.sendmail(mail_from, mail_to, msg.as_string())
    smtp.quit()
```
