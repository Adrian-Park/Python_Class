# 야구놀이

```python
import random
class BaseBall(object):
    def __init__(self,player=1,outcount=0,strike=0,ball=0):
        self.player = player
        self.outcount = outcount
        self.strike = strike
        self.ball = ball
        
        
    def count_player(self): #선수 교체시마다 선수 번호를 1 증가시킴
        self.player +=1
    
    
    def is_strike(self): #스트라이크나 볼을 발생시키는 메서드
        a=random.randint(0,1)
        if a == 0:
            print("스트라이크!!")
            self.strike += 1
        else:
            print("볼!!")
            self.ball += 1
        
    
    def get_status(self):
        print("{} 아웃, {} 스트라이크, {} 볼".format(self.outcount,self.strike,self.ball))
```

```python
player = BaseBall() #객체 생성
if __name__=='__main__':
    while player.outcount < 3:
        print("==== {}번째 선수 출격 ====\n".format(player.player))
        while True:
            print("공던짐 = > \n") 
            player.is_strike()
            player.get_status()
            print()
            if player.strike>=3:
                player.outcount+=1
                if player.outcount ==3:
                        print("3아웃")
                        print("쓰리아웃! 공수교체!\n")                        
                        player.strike = 0
                        player.ball =0
                        player.player = 1
                        break
                elif player.strike == 3:
                    print("=> {} 아웃! 공수 교체!\n".format(player.outcount))
                    player.strike = 0
                    player.ball =0
                    player.count_player()
                    break
            elif player.ball >=4:
                print("볼넷! 1루 출루\n")
                player.strike = 0
                player.ball =0
                player.count_player()
                break
```

# URL

```python
import requests

# GET
resp = requests.get('http://httpbin.org/get')
print(resp.text)
print(type(resp))

# POST
dic = {"id": 1, "name": "Kim", "age": 10}
resp = requests.post('http://httpbin.org/post', data=dic)
print(resp.text)
print(type(resp))

resp = requests.put('http://httpbin.org/put')
print(resp.text)
print(type(resp))

resp = requests.delete('http://httpbin.org/delete')
print(resp.text)
print(type(resp))
```

```python
import requests
from bs4 import BeautifulSoup
import re

movie_title = []
movie_point = []
movie_review = []

for n in range(1, 11):
    req = requests.get('https://movie.naver.com/movie/point/af/list.nhn?page='+str(n))
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.select('.movie')
    points = soup.select('td.title > div > em')
    reviews = soup.select('td.title')
    
    for dom in titles:
        movie_title.append(dom.txt)
    for dom in points:
        movie_point.append(dom.text)
    for dom in reviews:
        content = dom.contents[6]
        content = re.sub("[\n\t]", "", content)
        content = re.sub("신고", "", content)
        movie_review.append(content)
        
for i in range(len(movie_title)):
    print('영화제목:', movie_title[i])
    print('평점:', movie_point[i])
    print('리뷰:', movie_review[i])
```
